# Step 2: Data Preprocessing

`Correct for data imbalance`

Take only 1000 random samples from 'small' class, 1000 random samples from 'medium' class, and 1000 random samples from 'large' class

```python
# check for FPA_ID missing values
print(f"Number of missing valudes in FPA_ID: {df.FPA_ID.isna().sum()}")
```

```python
fpaid_fsize_small = df[df.FIRE_SIZE_LABEL == 'small'].FPA_ID
fpaid_fsize_medium = df[df.FIRE_SIZE_LABEL == 'medium'].FPA_ID
fpaid_fsize_large = df[df.FIRE_SIZE_LABEL == 'large'].FPA_ID
fpaid_fsize_verylarge = df[df.FIRE_SIZE_LABEL == 'very large'].FPA_ID
```

```python
np.random.seed(207)

# randomly draw 1000 FPA_ID from each classes
temp_fpaid_small = np.random.choice(
    fpaid_fsize_small,
    size = 1000,
    replace=False
)

temp_fpaid_medium = np.random.choice(
    fpaid_fsize_medium,
    size = 1000,
    replace=False
)

temp_fpaid_large = np.random.choice(
    fpaid_fsize_large,
    size = 1000,
    replace=False
)

temp_fpaid_verylarge = np.random.choice(
    fpaid_fsize_verylarge,
    size = 1000,
    replace=False
)

# retrieve dataframe for each classes
df_s = df[df.FPA_ID.isin(temp_fpaid_small)]
df_m = df[df.FPA_ID.isin(temp_fpaid_medium)]
df_l = df[df.FPA_ID.isin(temp_fpaid_large)]
df_vl = df[df.FPA_ID.isin(temp_fpaid_verylarge)]

# concatenate df for each class
df_mini = pd.concat([df_s, df_m, df_l, df_vl], axis=0)
df_mini = df_mini.reset_index(drop=True)
print(f"Shape of df_mini: {df_mini.shape}")
```

```python
df_mini.FIRE_SIZE_LABEL.value_counts()
```

```python
# recode fire_size_label  from string to numeric

firelabel_mapping = {
    'small': 0,
    'medium': 1,
    'large': 2,
    'very large': 3
}
print(f"{firelabel_mapping=}")

df_mini['FIRE_SIZE_LABEL'] = df_mini['FIRE_SIZE_LABEL'].map(firelabel_mapping)
```

`Shuffle the dataset`

```python
np.random.seed(207)

indices = list(range(df_mini.shape[0]))               # create a list of indices corresponding to rows in the df_mini dataset
shuffled_indices = np.random.permutation(indices)
df_mini = df_mini.reindex(index=shuffled_indices)     # change the ordering of the df_mini
```

`Create training, validation, and test datasets`

Using a 60/20/20 split

Approach: stratified random sampling, strata = FIRE_YEAR

```python
val_size = 0.2/(0.2 + 0.6)
df_train_main, df_test = train_test_split(df_mini, test_size= 0.2, random_state=207, stratify=df_mini.FIRE_YEAR)
df_train, df_val = train_test_split(df_train_main, test_size=val_size, random_state=207, stratify=df_train_main.FIRE_YEAR)

# print shapes
print(f"Shape of df_train: {df_train.shape}")
print(f"Shape of df_val: {df_val.shape}")
print(f"Shape of df_test: {df_test.shape}")
```

`Identify features of interest`

```python
# from FPA FOD data
trgt_feat_fod = ['FIRE_SIZE_LABEL', 'FIRE_YEAR', 'DISCOVERY_DOY', 'NWCG_CAUSE_CLASSIFICATION', 'LATITUDE', 'LONGITUDE']

# from GRIDMET: variables surround 5-day window centered on the fire discovery date
feat_gridmet = list(df_train.columns[df_train.columns.str.contains('_5D_')])

# from risk management assistance
feat_rmgmta = ['SDI']

# from fire stations
feat_firestation = ['No_FireStation_20.0km']

# from GACC
feat_gacc = ['GACC_PL']

# from global human modification
feat_ghm = ['GHM']

# from NDVI
feat_ndvi = ['NDVI-1day']      

# from national preparedness level
feat_npl = ['NPL']

# from social vulnerability index
feat_svi = ['EPL_PCI']

# from rangeland prod
feat_rangeland = ['rpms', 'rpms_1km']
```

```python
# compile selected vars
trgt_feat_selected = trgt_feat_fod + feat_gridmet + feat_rmgmta + feat_firestation + feat_gacc + feat_ghm + feat_ndvi + feat_npl + feat_svi + feat_rangeland
```

```python
# select only variables of interest (outcome and features)
df_train = df_train[trgt_feat_selected]
df_val = df_val[trgt_feat_selected]
df_test = df_test[trgt_feat_selected]

# print shapes
print(f"Shape of df_train: {df_train.shape}")
print(f"Shape of df_val: {df_val.shape}")
print(f"Shape of df_test: {df_test.shape}")
```

`Clean features`

Handle missing values

```python
# check for missing value
df_train.isna().sum().sort_values(ascending=False)
```

Approach to imputing missing values:
- GRIDMET variables: GRIDMET variables are climate variables. Variables selected contains "_5D_" in their column names and are numeric. We impute their missing values by using mean values.
- `GACC_PL`: It is a categorical variable, representing Geographic Area Coordination Center Prepareness Level. The value of `GACC_PL` is 1 for the least prepared level, and 5 for most prepared level. We impute missing values with 0, creating a new level representing missingness.
- `No_FireStation_20.0km`: It is a numeric variable, representing the number of fire stations in 20 km radius around the fire ignition point. We impute missing values with 0. That is, for missing values, we assume there is no fire station in the 20 km radius. 
- `SDI`: It is numeric variable, representing the relative difficulty of fire control. We impute its missing values by using the mean value of `SDI`.
- `EPL_PCI`: It is a numeric variable, representing percentile per capita income estimate. We impute its missing values by using the mean value of `EPL_PCI`.
- `NDVI-1day`: It is the NDVI (Normalized Difference Vegetation Index) on the day prior to ignition is a numeric variable, ranging from -1 to 1. Lower values indicate barren areas and higher values indicate greener vegetation. We imput its missing values by using the mean value of `NDVI-1day`.

```python
def impute_with_zero(df):
    """Impute the missing values in selected columns with zero

    Args:
     - df: a data frame

    Return:
    - df: a data frame 
    """

    # vars impute with zero
    cols_impute_zero = ['No_FireStation_20.0km', 'GACC_PL']

    for column in cols_impute_zero:
        df[column] = df[column].fillna(0)
    
    return df
```

```python
# selected vars: impute missing values with zero
df_train = impute_with_zero(df_train)
df_val = impute_with_zero(df_val)
df_test = impute_with_zero(df_test)
```

```python
# selected vars: impute with missing values the mean

# replace value of -999.0 (the only value less than 0) in EPL_PCI as missing values
for data in [df_train, df_val, df_test]:
    data['EPL_PCI'] = data['EPL_PCI'].replace(-999, np.nan)

cols_impute_mean = feat_gridmet + ['SDI', 'EPL_PCI', 'NDVI-1day']

imp_x = SimpleImputer(missing_values=np.nan, strategy='mean')
df_train[cols_impute_mean] = imp_x.fit(df_train[cols_impute_mean]).transform(df_train[cols_impute_mean])
df_val[cols_impute_mean] = imp_x.fit(df_train[cols_impute_mean]).transform(df_val[cols_impute_mean])
df_test[cols_impute_mean] = imp_x.fit(df_train[cols_impute_mean]).transform(df_test[cols_impute_mean])
```

```python
# sanity check for missing value 
df_train.isna().sum().sort_values(ascending=False)
```

Clean NWCG_CAUSE_CLASSIFICATION variable

```python
nwcg_cause_simplified = {
    'Human' : 'hum',
    'Natural' : 'nat',
    'Missing data/not specified/undetermined' : 'miss-unspec'
    
}

for data in [df_train, df_val, df_test]:
    data['NWCG_CAUSE_CLASSIFICATION'] = data['NWCG_CAUSE_CLASSIFICATION'].map(nwcg_cause_simplified)
```

```python
df_train['NWCG_CAUSE_CLASSIFICATION'].value_counts(dropna=False) # sanity check
```

`Add features`

One-hot-encoding

```python
cols_onehot = ['NWCG_CAUSE_CLASSIFICATION', 'GACC_PL']
df_train = pd.get_dummies(df_train, columns=cols_onehot)
df_val = pd.get_dummies(df_val, columns=cols_onehot)
df_test = pd.get_dummies(df_test, columns=cols_onehot)
```

`Define outcome and features`

```python
# define outcome
Y_train = df_train['FIRE_SIZE_LABEL']
Y_val = df_val['FIRE_SIZE_LABEL']
Y_test = df_test['FIRE_SIZE_LABEL']

print(f"Shape of Y_train: {Y_train.shape}")
print(f"Shape of Y_val: {Y_val.shape}")
print(f"Shape of Y_test: {Y_test.shape}")
```

```python
# define features
features = list(df_train.columns)
print(f"Number of items in features before removing outcome: {len(features)}")
features.remove('FIRE_SIZE_LABEL')
print(f"Number of items in features before removing outcome: {len(features)}")
print(f"{features=}")

X_train = df_train[features]
X_val = df_val[features]
X_test = df_test[features]

print(f"\nShape of X_train: {X_train.shape}")
print(f"Shape of X_val: {X_val.shape}")
print(f"Shape of X_test: {X_test.shape}")
```

`Standardized features`

```python
# exclude one-hot-encoded vars from later standardization
cols_ncwg = list(X_train.columns[X_train.columns.str.contains('NWCG_CAUSE_CLASSIFICATION')])
cols_gacc = list(X_train.columns[X_train.columns.str.contains('GACC_PL')])
cols_onehot = cols_ncwg + cols_gacc
cols_continuous = [item for item in list(X_train.columns) if item not in cols_onehot]

X_train_continuous = X_train[cols_continuous]
X_train_onehot = X_train[cols_onehot]
X_val_continuous = X_val[cols_continuous]
X_val_onehot = X_val[cols_onehot]
X_test_continuous = X_test[cols_continuous]
X_test_onehot = X_test[cols_onehot]

# standardized continuous features
sc_x = StandardScaler()

X_train_cont_std = sc_x.fit(X_train_continuous).transform(X_train_continuous)
X_val_cont_std = sc_x.fit(X_train_continuous).transform(X_val_continuous)
X_test_cont_std = sc_x.fit(X_train_continuous).transform(X_test_continuous)

# add one-hot-encoded vars back 
X_train_std = np.concatenate((X_train_cont_std, X_train_onehot), axis= 1)
X_val_std = np.concatenate((X_val_cont_std, X_val_onehot), axis= 1)
X_test_std = np.concatenate((X_test_cont_std, X_test_onehot), axis= 1)
```

# Step 3: Exploratory Data Analysis (EDA)

`correlation matrix of outcome and feature variables`

```python
corr = df_train.corr()                              # compute the correlation matrix
mask = np.triu(np.ones_like(corr, dtype=bool))      # generate a mask for the upper triangle
fig, ax = plt.subplots(figsize=(12,12))               # setup matplotlib figure
cmap = sns.diverging_palette(230,20, as_cmap=True)  # generate a custom diverging colormap
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, vmin=-1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},
            annot=True, fmt=".1f", annot_kws={"size": 5}, cbar=False)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=6) 
ax.set_yticklabels(ax.get_yticklabels(), fontsize=6) 
ax.set_title("Correlation Matrix of Features and Outcome in Training Data")
plt.tight_layout()
plt.show()
```

Note:

- From the correlation matrix heatmap, we observed that some features are redundant as they are highly correlated. The highest correlation values between features rounded to 1 decimal point are 1.0. In the heatmap, correlation values of those features are in red colored cells. Redundant features are prominent among GRIDMET features. 
- Some GRIDMET features, like fm1000_5D_min, rmax_5D_min, rmax_5D_max, seem useful for predicting the FIRE_SIZE_LABEL as their absolute correlation value are 0.5. NWCG_CAUSE_CLASSIFICATION_nat and NWCG_CAUSE_CLASSIFICATION_hum also seem useful for predicting FIRE_SIZE_LABEL. 
- The features that are likely to be least useful for predicting FIRE_SIZE_LABEL include pr_5D_min, GACC_PL_2.0, EPL_PCI since their correlation values with FIRE_SIZE_LABEL are close to zero.
